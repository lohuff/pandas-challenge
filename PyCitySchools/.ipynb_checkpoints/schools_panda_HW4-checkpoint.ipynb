```python
#import dependencies
import pandas as pd
```

```python
#add/define files and read them in
file_1 = "../Resources/schools_complete.csv"
file_2 = "../Resources/students_complete.csv"

school_data = pd.read_csv(file_1)
student_data = pd.read_csv(file_2)
```

```python
#merge the data into one set
total_data = pd.merge (student_data, school_data, how ="left", on = ["school_name", "school_name"])
total_data.head()
```

```python
#calculte total number of schools
total_schools = len(total_data["School ID"].unique())
total_schools
```

```python
#calculate total number of students
total_students = len(total_data["Student ID"].unique())
total_students
```

```python
#calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
#calculate the average math score
average_math_score = student_data["math_score"].mean()
average_math_score
```

```python
#calculate the average reading score
average_reading_score = student_data["reading_score"].mean()
average_reading_score
```

```python
#calculate the overall passing rate
overall_average_score = (average_math_score+ average_reading_score)/2
overall_average_score
```

```python
#calculate the % of students with a passing math score, 70 or higher
student_data["#passing_math"]= student_data["math_score"]>=70
passing_math = ((student_data["#passing_math"]).mean())*100
passing_math
```

```python
#Calculate the % of students with a passing reading score
student_data["#passing_reading"]= student_data["reading_score"]>=70
passing_reading = ((student_data["#passing_reading"]).mean())*100
passing_reading
```

```python
#calculate overall percentage passing
overall_passing = (passing_math + passing_reading)/2
overall_passing
```

```python
#create a district dataframe with results
district_results = [{"Total Schools": total_schools,
                    "Total Students": total_students,
                    "Total Budget": total_budget,
                    "Average Math Score": round(average_math_score,2),
                    "Average Reading Score": round(average_reading_score,2),
                    "Percent Passing Math": round(passing_math,2),
                    "Percent Passing Reading": round(passing_reading,2),
                    "Percent Overall Passing Rate": round(overall_passing,2)}]
district_df = pd.DataFrame(district_results)
```

```python
#format the df and display it
district_df["Percent Passing Math"]= district_df["Percent Passing Math"].map("{:,.2f}%".format)
district_df["Percent Passing Reading"]= district_df["Percent Passing Reading"].map("{:,.2f}%".format)
district_df["Percent Overall Passing Rate"]= district_df["Percent Overall Passing Rate"].map("{:,.2f}%".format)
district_df["Total Budget"] = district_df["Total Budget"].map("${:,.2f}".format)
district_df["Total Students"] = district_df["Total Students"].map("{:,}".format)

district_df
```

```python
#school data
total_data["passing_math"]=total_data["math_score"]>= 70
total_data["passing_reading"]= total_data["reading_score"]>= 70
total_data
```

```python
#groupby school name
school_name_group = total_data.groupby(["school_name"]).mean()
school_name_group["Per Student Budget"] = school_name_group["budget"]/school_name_group["size"]
school_name_group["Percent Passing Math"] =round(school_name_group["passing_math"]*100,2)
school_name_group["Percent Passing Reading"] =round(school_name_group["passing_reading"]*100,2)
school_name_group["Percent Overall Passing Rate"] = round(((school_name_group["passing_math"] + school_name_group["passing_reading"])/2)*100,3)
```

```python
#merge school data to collect new info
school_data_summary = pd.merge(school_name_group, school_data, how="left", on=["school_name", "school_name"])
del school_data_summary['size_y']
del school_data_summary['budget_y']
del school_data_summary['Student ID']
del school_data_summary['School ID_x']
```

```python
#create the dataframe
school_summary_df = pd.DataFrame({"School Name":  school_data_summary["school_name"],
                                "School Type": school_data_summary["type"],
                               "Total Students":school_data_summary["size_x"],
                               "Total School Budget": school_data_summary["budget_x"],
                               "Per Student Budget":school_data_summary["Per Student Budget"], 
                               "Average Math Score":round(school_data_summary["math_score"],2),
                               "Average Reading Score":round(school_data_summary["reading_score"],2), 
                               "Percent Passing Math": school_data_summary["Percent Passing Math"],
                               "Percent Passing Reading": school_data_summary["Percent Passing Reading"],
                               "Percent Overall Passing Rate": school_data_summary["Percent Overall Passing Rate"]}) 
```

```python
#format the dataframe and display it
school_summary_df["Total Students"] = school_summary_df["Total Students"].map("{:,.0f}".format)
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)

school_summary_df
```

```python
#sort and display the top 5 schools by overall passing rate
top_schools = school_summary_df.sort_values(["Percent Overall Passing Rate"], ascending=False)
top_schools.head()
```

```python
#sort and display the 5 worst performing schools by passing rate
bottom_schools = school_summary_df.sort_values(["Percent Overall Passing Rate"], ascending=True)
bottom_schools.head()
```

```python
#math scores by grade
#create a separate series per grade, grouping by school
ninth_grade= total_data[total_data["grade"] == "9th"].groupby("school_name").mean()["math_score"]
tenth_grade = total_data[total_data["grade"] == "10th"].groupby("school_name").mean()["math_score"]
eleventh_grade = total_data[total_data["grade"] == "11th"].groupby("school_name").mean()["math_score"]
twelfth_grade= total_data[total_data["grade"] == "12th"].groupby("school_name").mean()["math_score"]
```

```python
#make a new dataframe for math grade
math_grade_df = pd.DataFrame({"Ninth Grade":ninth_grade, "Tenth Grade":tenth_grade, 
                                     "Eleventh Grade":eleventh_grade, "Twelfth Grade":twelfth_grade})  
```

```python
#formatting and display
math_grade_df[["Ninth Grade","Tenth Grade","Eleventh Grade","Twelfth Grade"]] = math_grade_df[["Ninth Grade","Tenth Grade","Eleventh Grade","Twelfth Grade"]].applymap("{:.2f}".format)
math_grade_df
```

```python
#perform previous operations, but for reading scores
ninth_grade= total_data[total_data["grade"] == "9th"].groupby("school_name").mean()["reading_score"]
tenth_grade = total_data[total_data["grade"] == "10th"].groupby("school_name").mean()["reading_score"]
eleventh_grade = total_data[total_data["grade"] == "11th"].groupby("school_name").mean()["reading_score"]
twelfth_grade= total_data[total_data["grade"] == "12th"].groupby("school_name").mean()["reading_score"]

reading_grade_df = pd.DataFrame({"Ninth Grade":ninth_grade, "Tenth Grade":tenth_grade, 
                                     "Eleventh Grade":eleventh_grade, "Twelfth Grade":twelfth_grade})  

reading_grade_df[["Ninth Grade","Tenth Grade","Eleventh Grade","Twelfth Grade"]] = reading_grade_df[["Ninth Grade","Tenth Grade","Eleventh Grade","Twelfth Grade"]].applymap("{:.2f}".format)
reading_grade_df
```

```python
#scores by school spending

spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_data_summary["Spending Ranges (Per Student)"] = pd.cut(school_data_summary["Per Student Budget"], spending_bins, labels=labels)

school_spending_group = school_data_summary.groupby("Spending Ranges (Per Student)").mean() 
```

```python
#remove unwanted columns and display
del school_spending_group['size_x']
del school_spending_group['budget_x']
del school_spending_group['Per Student Budget']
del school_spending_group['School ID_y']
del school_spending_group['passing_math']
del school_spending_group['passing_reading']

school_spending_group
```

```python
#mean scores per spending range
spending_math_scores = school_spending_group.groupby(["Spending Ranges (Per Student)"]).mean()["math_score"]
spending_reading_scores = school_spending_group.groupby(["Spending Ranges (Per Student)"]).mean()["reading_score"]
spending_passing_math = school_spending_group.groupby(["Spending Ranges (Per Student)"]).mean()["Percent Passing Math"]
spending_passing_reading = school_spending_group.groupby(["Spending Ranges (Per Student)"]).mean()["Percent Passing Reading"]
overall_passing_spending = school_spending_group.groupby(["Spending Ranges (Per Student)"]).mean()["Percent Overall Passing Rate"]
```

```python
#create spending summary dataframe
spending_summary_df = pd.DataFrame({"Average Math Score":spending_math_scores, "Average Reading Score":spending_reading_scores, 
                                     "Percent Passing Math":spending_passing_math, "Percent Passing Reading":spending_passing_reading, 
                                    "Percent Overall Passing Rate": overall_passing_spending}) 
spending_summary_df
```

```python
#scores by school size
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_data_summary["School Size"] = pd.cut(school_data_summary["size_x"], size_bins, labels=labels)
school_data_summary
```

```python
size_summary = school_data_summary.groupby("School Size").mean() 
size_summary
```

```python
#formatting and display
del size_summary['budget_x']
del size_summary['Per Student Budget']
del size_summary['School ID_y']
del size_summary['passing_math']
del size_summary['passing_reading']
size_summary
```
